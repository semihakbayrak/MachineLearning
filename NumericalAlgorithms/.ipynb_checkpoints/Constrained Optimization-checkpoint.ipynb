The most basic constrained optimization type is:
$$
\text{minimize } f(x) \\
\text{subject to } g(x) = 0
$$

An example of it is visualized above. Since we are constrained within the set of numbers that satisfy $g(x) = 0$, steps we take must be perpendicular to gradient $\nabla g$ so that $g(x)$ don't change. This can be seen when we write Taylor expansion of $g(x)$:
$$
g(\hat{x}+\Delta \hat{x}) \approx g(\hat{x}) + \Delta \hat{x} \nabla g(x) |_{\hat{x}} + \mathcal{O}(x^2)
$$
$g(\hat{x}+\Delta \hat{x}) = 0$, $g(\hat{x})=0$, then $\nabla g(x) |_{\hat{x}}$ must be 0.

In this scenerio we will not reach the minimum of $f(x)$ indicated by star. But we are looking for such a point that $\nabla f$ will be perpendicular to infinitesimal $\Delta x$ which means that there won't be any component of $\nabla f$ on $\Delta x$. So, the infinitesimal step $\Delta x$ can not change $f(x)$ further. Or in another words, we seek for such a point that $f(x)$ can not be improved further on the constrain function. At this point $\nabla f$ and $\nabla g$ are parallel to each other. Thus, we can write that $\nabla f = \lambda \nabla g$. Equivalently, $\nabla f - \lambda \nabla g = 0$. We can reformalize the problem with introducing the following Langragian cost function $\Lambda(x,\lambda) = f(x) - \lambda g(x)$. Note that we seek for $\nabla_{x} \Lambda(x,\lambda) = \nabla f(x) - \lambda \nabla g(x) = 0$. Also note that $\nabla_{\lambda} \Lambda(x,\lambda) = -\nabla g(x) = 0$ must be satisfied.
